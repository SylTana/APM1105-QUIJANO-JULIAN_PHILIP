```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import tabulate as t

movies_df = pd.read_csv(r"C:\Users\Julian Quijano\Desktop\CSV\movies.csv")
ratings_df = pd.read_csv(r"C:\Users\Julian Quijano\Desktop\CSV\ratings.csv")

#print(movies_df['title'].head(10))
print('\n')

# 1. How many movies are released each year? What is the percentage of change over time?

movies_df['year'] = movies_df['title'].str.extract(r'\((\d{4})\)').fillna(0).astype(int)

frequency = movies_df['year'].value_counts().sort_values(ascending=True).reset_index(name='frequency')

for i, j in frequency.iterrows():
    if j['year'] == 0:
        frequency = frequency.drop(i)

frequency = frequency.sort_values('year', ascending=True)
frequency = frequency.rename(columns={'index': 'year'})

print(t.tabulate(frequency, headers=['Year', 'Frequency'], tablefmt='fmt', showindex=False))
print('\n')
plot.figure(figsize=(15, 10))
plot.bar(frequency['year'], frequency['frequency'], color='purple')
plot.xlabel('Year')
plot.ylabel('Frequency')
plot.title('No. of Movies Released from 1874 to 2019')
plot.show()

percent_change = pd.DataFrame()
percent_change['Year'] = frequency['year']
percent_change['Change'] = frequency['frequency'].pct_change()
percent_change['Change'] = round(percent_change['Change'] * 100, 2)

print('Percentage of Change over Time')
print(t.tabulate(percent_change, headers=['Year', 'Percent Change'], tablefmt='presto', showindex=False))
print('\n')
percent_change_graph = np.diff(percent_change['Change'])

figure_2, ax = plot.subplots(figsize=(15, 10))
ax.plot(percent_change['Year'], percent_change['Change'], color='purple', linewidth=4)
ax.set(title='Percent Change over Time', xlabel='Year', ylabel='Percent Change', ylim=-100)
ax.grid(linestyle='solid')
plot.show()
```

```python
# 2. What are the most popular genre of movies?

genres = movies_df['genres'].str.split('|', expand=True).stack().reset_index(level=1, drop=True)
genre_count = genres.value_counts(sort=False)
genre_count = genre_count.to_frame().reset_index()
genre_count.columns = ['Genre', 'Count']
genre_count = genre_count.sort_values(['Genre'], ascending=True).reset_index(drop=True)

print(t.tabulate(genre_count, headers=['Genre', 'Total'], tablefmt='fmt', showindex=False))
print("The most popular genre of movies is Dramas with", max(genre_count['Count']), "amount of movies.")
print('\n')

figure_3 = plot.figure(figsize=(15, 8))
plot.rcParams['font.size'] = 6
plot.bar(genre_count['Genre'], genre_count['Count'], color='purple')
plot.xlabel('Genre')
plot.ylabel('Total')
plot.title('Frequency of Movie Genres')
plot.xticks(rotation=90)
plot.show()

# 3. Does the taste of movies change over time?

genre_freq_year = pd.DataFrame()
genre_freq_year['Year'] = movies_df['year']
genre_freq_year['Genres'] = movies_df['genres']

genre_freq_year = genre_freq_year.assign(Genres=genre_freq_year.Genres.str.split('|')).explode('Genres')
genre_freq_year = genre_freq_year[genre_freq_year.Genres.ne('(no genres listed)') & genre_freq_year.Genres.ne('')]
genre_freq_year = genre_freq_year.reset_index(drop=True)

genre_freq_year['Frequency'] = 1
genre_freq_year = genre_freq_year.groupby(['Year', 'Genres']).sum().reset_index()
for index, row in genre_freq_year.iterrows():
    if row['Year'] == 0:
        genre_freq_year = genre_freq_year.drop(index)

genre_years = genre_freq_year.groupby(['Genres', 'Year']).sum()
genre_years = genre_years.unstack().fillna(0)
genre_years['Frequency'] = genre_years['Frequency'].astype(int)
print(genre_years)

figure_4, ax = plot.subplots(figsize=(10, 5))

for i, j in genre_years.iterrows():
    genre_graph = j['Frequency']
    genre_graph.plot(ax=ax, xlabel='Year', ylabel='Frequencies', title="Number of Different Genres from 1870 to 2020", xticks=range(1870, 2030, 10))

ax.legend(prop={'size': 7})
plot.grid(linestyle='solid')
plot.show()
print("Yes, movie tastes definitely change over the years.\n")

# 4. What are the top 20 movies that has the top amount of users rating and has the highest rate?

ratings_df = ratings_df.drop(['timestamp'], axis=1)
ratings_count_per_movie = ratings_df['movieId'].value_counts().head(20).head(20).tolist()
id_for_rcpm = ratings_df['movieId'].value_counts().head(20).index.tolist()

movieslist = []
for i in id_for_rcpm:
    for j in movies_df['movieId']:
        if i == j:
            movieslist.append(movies_df['title'].iloc[movies_df.index[i]])

movie_and_ratings = []
for i in range(20):
    movie_and_ratings.append([movieslist[i], ratings_count_per_movie[i]])

ratings_and_movies = pd.DataFrame(movie_and_ratings)
ratings_and_movies.columns = ['Title', 'Amount of User Ratings']
print(ratings_and_movies)

figure_5 = plot.figure(figsize=(15, 8))
plot.rcParams['font.size'] = 4
plot.bar(movieslist, ratings_count_per_movie, color='purple')
plot.xlabel('Movie')
plot.ylabel('Total User Ratings')
plot.title('Top 20 Movies based on Total User Ratings')
plot.xticks(rotation=20)
plot.show()

#highest rating
t_ratings = ratings_df.groupby(['movieId']).mean().sort_values(['rating'], ascending=False).head(20).head(20)
id_for_t_ratings = ratings_df['movieId'].value_counts().head(20).index.tolist()
t_list = t_ratings['rating'].tolist()

movieslist2 = []
for i in id_for_t_ratings:
    for j in movies_df['movieId']:
        if i == j:
            movieslist2.append(movies_df['title'].iloc[movies_df.index[i]])

movie_and_totalratings = []
for i in range(20):
    movie_and_totalratings.append([movieslist2[i], t_list[i]])

print('\nTop 20 Movies by Average Ratings')
print(t.tabulate(movie_and_totalratings, headers=['Movie', 'Average Rating'], tablefmt='presto', showindex=False))

# 5. What is the most popular season for releasing movies?
```
