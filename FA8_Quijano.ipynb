## Create an exponent calculator using both iterations and recursion, with rendering time.

```python
import time


def iterative_exponent(x, y):
    answer = 1
    for i in range(0, y):
        answer *= x
    return answer


def recursive_exponent(x, y):
    if y == 0:
        return 1
    else:
        return x * recursive_exponent(x, y - 1)

# Executing
base = int(input("Enter a value for the base: "))
exponent = int(input("Enter a value for the exponent: "))

print("\nIterative function:")
start_time = time.time()
print(base, "raised to", exponent,  "=", iterative_exponent(base, exponent))
print("Total time elapsed = ", time.time()-start_time)

print("\nRecursive function:")
start_time = time.time()
print(base, "raised to", exponent,  "=", recursive_exponent(base, exponent))
print("Total time elapsed = ", time.time()-start_time)
```

##  Iteratively and recursively calculate the sum of the first n powers of 2, along with the rendering time of each function using the time module.

```python
import time

def iterative_sum(x):
    total_sum = 0
    exponentt = 1
    for i in range(0, x):
        exponentt *= 2
        total_sum += exponentt
    return total_sum


def recursive_sum(x, total_sum=0, exponent=1):
    if x == 0:
        return total_sum
    else:
        exponent *= 2
        total_sum += exponent
        return recursive_sum(x - 1, total_sum, exponent)

# Executing
exponent = int(input("\n\nEnter a value for the exponent: "))

print("\nIterative function:")
start_time = time.time()
print("The sum of the first", exponent, "powers of 2 is", iterative_sum(exponent))
print("Total time elapsed = ", time.time()-start_time)


print("\nRecursive function:")
start_time = time.time()
print("The sum of the first", exponent, "powers of 2 is", recursive_sum(exponent))
print("Total time elapsed = ", time.time()-start_time)
```

## Recursive functions in both programs are faster since they only have to call their own function and not call many variables to execute.
