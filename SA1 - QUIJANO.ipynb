```python
import requests
from collections import Counter

# URL of the raw dataset file on GitHub
url = 'https://raw.githubusercontent.com/AllenDowney/ThinkPython2/master/code/words.txt'

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Save the content of the response to a file
    with open('words.txt', 'wb') as f:
        f.write(response.content)
        print('Dataset saved to file.')
else:
    print('Failed to download dataset.')

# Print File
main_file = response.text.split()
# print(main_file)

# Total number of words in the file
print("\nTotal number of words: ", int(len(main_file)))

# Longest word in the file
word = 0
count = 0
# Set count 0
for i in main_file:  # Go through the whole list
    if len(i) > count:  # Checking for the longest word(string)
        count = len(i)
        word = i

print("\nThe longest word: " + word)

# Print the number of words that start with a certain letter (use user input)
letter = input("\nChoose a letter: ").lower()
letter_count = 0
for i in main_file:
    if i[0] == letter:
        letter_count += 1

print("There are " + str(letter_count) + " words which begin with this letter.")

# Print all the words that contain a certain substring (use user input).
substr = input("\nInput a specific substring: ")
i_count = 0
substr_list = []
for i in main_file:
    if i.find(substr) != -1:
        substr_list += i.split()
        i_count += 1

if i_count == 0:
    print("There were no words that contain this substring.\n")
elif i_count == len(substr_list):
    print("These are the words that contain this substring:")
    print(substr_list, "\n")

# Print all the words that are palindromes (i.e., read the same backwards and forwards).
palindromes = []
for i in main_file:
    if i[::-1] == i:
        palindromes += i.split()

print("These are the palindromes in this list:")
print(palindromes, "\n")

# Count the frequency of each letter.
ignore = ['\n', '\r']
freq_letter = Counter(response.text)
for letter2 in list(freq_letter):
    if letter2 in ignore:
        del freq_letter[letter2]

print("These are the frequencies of each letter:")
print(sorted(freq_letter.items()))
```

```python
import requests
import random


url = 'https://raw.githubusercontent.com/AllenDowney/ThinkPython2/master/code/words.txt'

response = requests.get(url)
main_file = response.text.split()

def randomword(): # Choosing a random word.
    word = (random.choice(main_file))
    return word

def guessing(): # Taking the user's guess.
    guess = input("Take a guess: ").lower()
    if len(guess) != 1 or not guess.isalpha():  # Checking if guess is a single letter.
        print("\nSorry, this is not a valid guess.")
        del guess
        guess = input("Take another guess: ").lower()
        if len(guess) != 1 or not guess.isalpha():  # Checking if guess is a single letter.
            print("\nSorry, this is not a valid guess.")
            del guess
            guess = input("Take another guess: ").lower()
            if len(guess) != 1 or not guess.isalpha():  # Checking if guess is a single letter.
                print("\nSorry, this is not a valid guess.")
                del guess
                guess = input("Take another guess: ").lower()
                if len(guess) != 1 or not guess.isalpha():  # Checking if guess is a single letter.
                    print("\nYou're not playing the game.")
                    return None
            else:
                return guess
        else:
            return guess
    else:
        return guess

def printing(word, guessed): # Printing the user's guess.
    answer = ""

    for letter in word:
        if letter in guessed:
            answer += letter + " "
        else:
            answer += "_ "
    print(answer)

def game():
    word = randomword()
    attempts = len(word)
    guessed = []
    print("Welcome!")
    print("You are guessing a {}-letter word.".format(attempts))

    while attempts > 0:
        guess = guessing()

        guessed.append(guess)  # Joining in a list, all guesses.

        printing(word, guessed)

        if None or attempts == 0:
            print("I'm sorry, you've run out of attempts. The answer is \033[1m{}\033[0m.".format(word))
            return False
        elif guess in word:
            print("Correct! Next!\n")
        else:
            print("Try again!")
            attempts -= 1
            print("You have {} attempts left.\n".format(attempts))



        if all(letter in guessed for letter in word):
            print("Amazing! You guessed the word! Congratulations!")
            return False
        if attempts == 0:
            print("I'm sorry, you've run out of attempts. The answer is \033[1m{}\033[0m.".format(word))
            return False

# main()
for i in range(0, 2):    
    while True:
        game()
        print("Thank you for playing! Goodbye.")
        break
```
